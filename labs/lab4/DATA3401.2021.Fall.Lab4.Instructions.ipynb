## Exercise 9: 
Use all of the previous exercises to implement a full tic-tac-toe game, where an appropriate board is drawn, 2 players are repeatedly asked for a location coordinates of where they wish to place a mark, and the game status is checked until a player wins or a draw occurs.

````python
# Write you solution here
# Write you solution here
# Write you solution here
# Write you solution here
import numpy as np
def board_create(n):
    board = list()
    for i in range(n):
        row = list()
        for j in range(n):
            row.append(' ')
        board.append(row)
    return board
def printBoard(board):
    for row in board:
        print("--- --- ---")
        for j in range(len(row)):
            print("|",row[j], end ="")
        print("\n")
def play_game(theboard):
    turn = 'X'
    count = 0
    
    for i in range(n*n+1):
        printBoard(theboard)
        print("Its the turn for: "+turn)
        move = list(input("Enter the position you wanna enter value in the form of A1 or B2"))
        if move[0]=='A':
            move[0]=0
        elif move[0]=='B':
            move[0]=1
        elif move[0]=='C':
            move[0]=2
        elif move[0]=='D':
            move[0]=3
        elif move[0]=='E':
            move[0]=4
        elif move[0]=='F':
            move[0]=5
        move[1]=int(int(move[1])-1)
        #print(move)
        if theboard[move[0]][move[1]] == ' ':
            theboard[move[0]][move[1]] = turn
            count += 1
        else:
            print("That place is already filled.\nMove to which place?")
            i -= 1
            continue
        if count >=5:
            for row in theboard:
                if len(set(row))==1 and row[0]!=' ':
                    print("Game Over")
                    printBoard(theboard)
                    return row[0]
            newboard = np.transpose(theboard)
            for column in newboard:
                if len(set(column)) ==1 and column[0]!=' ':
                        print("Game Over")
                        printBoard(theboard)
                        return column[0]
            if len(set([theboard[i][i] for i in range(len(theboard))])) == 1 and theboard[0][0]!=' ':
                print("game over")
                printBoard(theboard)
                return theboard[0][0]
            elif (len(set([theboard[i][len(theboard)-i-1] for i in range(len(theboard))])) == 1 and theboard[0][len(theboard)-1]!=' '):
                print("Game Over")
                printBoard(theboard)
                return theboard[0][len(theboard)-1]
        
        if count == n*n:
            print("Game over")
            printBoard(theboard)
            return -1
            break
        if turn =='X':
            turn = 'O'
        else:
            turn = 'X'
````

````python
# Test your solution here
n=3
board = board_create(n)
ret_val = play_game(board)
if ret_val == -1:
    print("There's a tie")
else:
    print("The winner is  "+str(ret_val))
    
````
