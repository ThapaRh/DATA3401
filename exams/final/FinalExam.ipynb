If you invest a principal value of $P$ at time 0, and interest is continuously compounded at a rate $r$ ($0<r<1$), then the amount of money you would have after $t$ years is
$$M(t) = Pe^{rt}.$$
Fry has \\$0.01 in his bank account, but is accidentally transported 1,000 years into the future. He returns to his bank (which luckily still exists) to see how much money he now has. Assume Fry's account earns continuously compounded interest at a rate of 5\% (or $r=0.05$).

1. Create a new numpy array that calculates how much money Fry's account has at each year
1. Plot Fry's money over the given timeframe

Now if Fry only earned simple interest, the amount of money he would have after $t$ years would be
$$ S(t) = P(1+rt).$$

1. Perform the same procedure as steps 1-3 above assuming Fry earns 5\% simple interest.
1. To illustrate the difference, also plot $M(t)-S(t)$ over the timeframe.
1. What do you conclude?

*Aside:* This problem, while silly, should teach you an important lesson: invest your money as early as you possibly can!

```python
# Write your code here
import math
import pandas as pd
import matplotlib.pyplot as plt 
Principal = 0.01
r = 0.05
#array of increasing time
y= []
i=0
while i<1000:
    y.append(i+1)
    i=i+1
#array of money he has every year
total_money = []
j = 0
while j<1000:
    money_ey= Principal*math.exp(r*(j+1))
    Principal = money_ey
    total_money.append(money_ey)
    j=j+1
#with SI
SM =[]
k=0
while k<1000:
    money_ey= Principal*(1+(r*(k+1)))
    Principal = money_ey
    SM.append(money_ey)
    k=k+1

plt.plot(y, total_money, color= 'green' ,label = "C_I")
plt.show()
plt.plot(y, SM, color= 'green' ,label = "SI")
plt.show()


```

```python
# Write your code here
# Write your code here
import numpy as np
import random
import matplotlib.pyplot as plt


def random_walk(num_steps, init_position):
    """This one is with equal probability"""
    
    walk_choices = [np.array(arr) for arr in [(0, 1), (1, 0), (0, -1), (-1, 0)]]
    
    final_out_x = [init_position[0]]
    final_out_y = [init_position[1]]

    last_coordinate = init_position
    
    for _ in range(num_steps):
        last_coordinate = last_coordinate + random.choice(walk_choices)
        final_out_x.append(last_coordinate[0])
        final_out_y.append(last_coordinate[1])
        
    return final_out_x, final_out_y


def random_walk_different_probability(num_steps, init_position):
    """This one is with different probability"""
    
    walk_choices = [np.array(arr) for arr in [*[(0, 1) for _ in range(7)], (1, 0), (0, -1), (-1, 0)]]
    
    final_out_x = [init_position[0]]
    final_out_y = [init_position[1]]

    last_coordinate = init_position
    
    for _ in range(num_steps):
        last_coordinate = last_coordinate + random.choice(walk_choices)
        final_out_x.append(last_coordinate[0])
        final_out_y.append(last_coordinate[1])
    
    return final_out_x, final_out_y



x, y = random_walk(10000, np.array([0, 0]))
plt.plot(x, y)

```

```python
import random


class Card:
    def __init__(self, value: str, suit: str):
        """provideet value and suit attribute"""
        self.value = value
        self.suit = suit

    def __repr__(self):
        return f"{self.suit}-{self.value}"
        

class Deck:
    CARD_VALUES = ["A", "K", "Q", "J", *[str(i) for i in range(10, 1, -1)]]
    CARD_SUITS = ["Spades", "Clubs", "Diamonds", "Hearts"]

    def __init__(self, shuffled: bool):
        self.shuffled = shuffled
        self.cards = []

        self._create_cards()
        
    def _create_cards(self) -> None:
        for card_suit in self.CARD_SUITS:        
            for card_value in self.CARD_VALUES:
                self.cards.append(Card(
                    value=card_value,
                    suit=card_suit
                ))

        if self.shuffled is True:
            self.shuffle()

    def shuffle(self) -> None:
        random.shuffle(self.cards)

    def deal_cards(self, num_cards: int) -> list[Card]:
        dealt_cards = self.cards[0:num_cards]
        self.cards = self.cards[num_cards:]

        return dealt_cards


class Hand(Deck):
    def __init__(self, num_cards: int, shuffled: bool=True):
        super().__init__(shuffled=shuffled)
        
        self.num_cards = num_cards
        self.hand = []

        self.set_new_hand()

    def set_new_hand(self):
        self.hand = self.deal_cards(num_cards=self.num_cards)

    def get_suit_if_flush(self):
        first_card = self.hand[0].suit

        for card in self.hand:
            if first_card != card.suit:
                return None

        return card.suit

    def is_straight_hand(self):
        hand_values = set([card.value for card in self.hand])
        if len(hand_values) != self.num_cards:
            return False

        ordered_card_value_str = "|".join(self.CARD_VALUES)
        ordered_hand_value_list = []

        for card_value in self.CARD_VALUES:
            if card_value in hand_values:
                ordered_hand_value_list.append(card_value)
        
        ordered_hand_value_str = "|".join(ordered_hand_value_list)

        if ordered_hand_value_str in ordered_card_value_str:
            return True
        else:
            return False


def print_result(hand: Hand, number_of_hands: int):
    for i in range(number_of_hands):
        flush_data = hand.get_suit_if_flush()
        is_straight = hand.is_straight_hand()
        
        print(f"HAND {i + 1}: {hand.hand}")
        print(f"1. {'Not a Flush' if flush_data is None else f'Is {flush_data} Flush'}")
        print(f"2. {'Is Straight' if is_straight is True else 'Is Not Straight'}")
        hand.set_new_hand()


def create_unshuffeled_deck():
    unshuffeled_deck_hand = Hand(num_cards=5, shuffled=False)
    print_result(hand=unshuffeled_deck_hand, number_of_hands=5)


def create_shuffeled_deck():
    unshuffeled_deck_hand = Hand(num_cards=5, shuffled=True)
    print_result(hand=unshuffeled_deck_hand, number_of_hands=9)


if __name__ == "__main__":
    create_shuffeled_deck()
```
